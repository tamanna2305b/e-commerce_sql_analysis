SQL E-COMMERCE ANALYSIS

Executive Summary

This project analyses an e-commerce sales dataset using SQL to answer key business questions related to revenue, customer behaviour, product performance and operational efficiency.

The objective is to demonstrate SQL skills by solving real-world business problems using filtering, aggregation, grouping and advanced analytical queries. Each query is followed by business insights and recommendations based on the results.


Tools Used:
- SQL (SQLite)
- Python
- Pandas
- Google Colab


```python
import sqlite3
import pandas as pd
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_csv("ecommerce_orders_dataset.csv")

df.head()
```

```python
conn = sqlite3.connect(":memory:")
```

```python
df.to_sql("orders", conn, index=False, if_exists="replace")
```

```python
query = """
SELECT COUNT(*) AS Total_Orders
FROM orders;
"""

pd.read_sql(query, conn)
```

Insight-

This query calculates the total number of orders processed by the business. It provides a baseline measure of sales activity and overall business scale.

Idea-

Tracking order volume over time to identify seasonal trends and evaluate the effectiveness of marketing campaigns and promotional activities.

```python
query = """
SELECT ROUND(SUM(Order_Amount),2) AS Total_Revenue
FROM orders;
"""

pd.read_sql(query, conn)
```

Insight-

This query calculates the total revenue generated from all completed orders. It provides an overview of the company's sales performance.

Idea-

Monitoring total revenue regularly and compare it across different months or years to identify growth trends and evaluate business performance.

```python
query = """
SELECT ROUND(AVG(Order_Amount),2) AS Average_Order_Value
FROM orders;
"""

pd.read_sql(query, conn)
```

Insight-

The average order value measures how much customers spend per transaction. It is a key indicator of customer purchasing behaviour.

Recommendation-

Increase the average order value through product bundles, cross-selling and personalised recommendations.

```python
query = """
SELECT COUNT(DISTINCT Customer_ID) AS Total_Customers
FROM orders;
"""

pd.read_sql(query, conn)
```

Insight-

This query identifies the total number of unique customers who have placed orders.

 Recommendation-

Compare customer growth over time to understand acquisition trends and customer retention.

```python
query = """
SELECT
    Country,
    ROUND(SUM(Order_Amount),2) AS Revenue
FROM orders
GROUP BY Country
ORDER BY Revenue DESC;
"""

pd.read_sql(query, conn)
```

 Insight-

This query identifies the countries that contribute the highest revenue.

 Recommendation-

Focus marketing efforts and inventory planning on high-performing countries while investigating opportunities to improve sales in lower-performing markets.

```python
query = """
SELECT
    City,
    ROUND(SUM(Order_Amount),2) AS Total_Revenue
FROM orders
GROUP BY City
ORDER BY Total_Revenue DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

Key Takeaways

The first few queries give a good overview of the business. The dataset contains a large number of orders and customers, giving enough data to analyse sales trends. Looking at total revenue and average order value also gives a useful starting point before exploring which customers, products and regions contribute the most to the business.

 Insight-

This query identifies the ten cities generating the highest revenue. These locations represent the company's strongest markets.

 Recommendation-

Increase marketing investment and inventory availability in these high-performing cities while analysing why lower-performing cities generate less revenue.

```python
query = """
SELECT
    Product_Category,
    ROUND(SUM(Order_Amount),2) AS Revenue
FROM orders
GROUP BY Product_Category
ORDER BY Revenue DESC;
"""

pd.read_sql(query, conn)
```

Insight-

This analysis shows which product categories contribute the most revenue.

Recommendation-

Prioritise high-performing categories for promotions and product expansion while reviewing strategies for lower-performing categories.

```python
query = """
SELECT
    Customer_Segment,
    ROUND(SUM(Order_Amount),2) AS Revenue
FROM orders
GROUP BY Customer_Segment
ORDER BY Revenue DESC;
"""

pd.read_sql(query, conn)
```

 Insight-

Different customer segments contribute differently to overall revenue.

Recommendation-

Develop targeted marketing campaigns for each customer segment to maximise sales and customer retention.

```python
query = """
SELECT
    Payment_Method,
    ROUND(SUM(Order_Amount),2) AS Revenue
FROM orders
GROUP BY Payment_Method
ORDER BY Revenue DESC;
"""

pd.read_sql(query, conn)
```

 Insight-

This analysis highlights which payment methods customers prefer.

 Recommendation-

Ensure the most popular payment methods provide a seamless checkout experience while encouraging adoption of other secure payment options.

```python
query = """
SELECT
    Traffic_Source,
    ROUND(SUM(Order_Amount),2) AS Revenue
FROM orders
GROUP BY Traffic_Source
ORDER BY Revenue DESC;
"""

pd.read_sql(query, conn)
```

Insight-

This query evaluates which marketing channels generate the highest revenue.

 Recommendation-

Allocate more marketing budget towards high-performing traffic sources while optimising or reconsidering less effective channels.

Key Takeaways

 The analysis now shows where the business performs best. Some cities and product categories contribute much more revenue than others, while customer segments and payment methods reveal differences in purchasing behaviour. Looking at delivery times and return rates also highlights areas where the company could improve the customer experience.

```python
query = """
SELECT
    Product_Category,
    COUNT(*) AS Total_Orders,
    SUM(CASE WHEN Returned = 'Yes' THEN 1 ELSE 0 END) AS Returned_Orders,
    ROUND(
        100.0 * SUM(CASE WHEN Returned = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) AS Return_Rate_Percent
FROM orders
GROUP BY Product_Category
ORDER BY Return_Rate_Percent DESC;
"""

pd.read_sql(query, conn)
```

 Insight-

This query measures the percentage of returned orders for each product category.

Recommendation-

Investigate categories with high return rates to identify quality issues, inaccurate product descriptions or delivery-related problems.

```python
query = """
SELECT
    Customer_ID,
    ROUND(SUM(Order_Amount),2) AS Total_Spent
FROM orders
GROUP BY Customer_ID
ORDER BY Total_Spent DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

 Insight-

This query identifies the company's highest-spending customers.

 Recommendation-

Reward these customers with loyalty programmes, exclusive offers and personalised promotions to improve retention.

```python
query = """
SELECT
    Product_Category,
    ROUND(SUM(Profit_Amount),2) AS Total_Profit
FROM orders
GROUP BY Product_Category
ORDER BY Total_Profit DESC;
"""

pd.read_sql(query, conn)
```

 Insight-

Revenue does not always equal profit. This analysis identifies which product categories contribute the most profit.

Recommendation-

Focus future investment and inventory on high-profit categories while reviewing pricing strategies for less profitable products.

```python
query = """
SELECT
    Customer_ID,
    ROUND(SUM(Order_Amount),2) AS Total_Spent,
    RANK() OVER (
        ORDER BY SUM(Order_Amount) DESC
    ) AS Customer_Rank
FROM orders
GROUP BY Customer_ID;
"""

pd.read_sql(query, conn)
```

Insight-

This query ranks customers based on total spending, helping identify the most valuable customers.

Recommendation-

Use customer rankings to create VIP programmes and targeted retention strategies.

```python
query = """
SELECT
    Order_Date,
    ROUND(Order_Amount,2) AS Revenue,
    ROUND(
        SUM(Order_Amount) OVER(
            ORDER BY Order_Date
        ),
        2
    ) AS Running_Total
FROM orders
ORDER BY Order_Date;
"""

pd.read_sql(query, conn)
```

Insight-

A running total shows cumulative revenue over time, making it easier to monitor business growth.

Recommendation-

Use cumulative revenue tracking to compare growth against sales targets and identify periods of slower performance.

```python
query = """
SELECT
    strftime('%Y-%m', Order_Date) AS Month,
    ROUND(SUM(Order_Amount),2) AS Revenue
FROM orders
GROUP BY Month
ORDER BY Month;
"""

pd.read_sql(query, conn)
```

Insight-

This query shows how revenue changes from month to month.

Recommendation-

Identify seasonal peaks and low-demand periods to improve inventory planning and marketing campaigns.

```python
query = """
SELECT
    Product_Category,
    ROUND(AVG(Review_Rating),2) AS Average_Rating
FROM orders
GROUP BY Product_Category
ORDER BY Average_Rating DESC;
"""

pd.read_sql(query, conn)
```

 Insight-

Customer satisfaction varies across product categories.

Recommendation-

Investigate categories with lower ratings to improve product quality and customer experience.

 Key Takeaways-

The analysis focuses on customer value and long-term business performance. Ranking customers helps identify the most valuable buyers, while profit and monthly revenue trends give a better picture of how the business is performing over time. These insights can support decisions around marketing, inventory planning and customer retention.
